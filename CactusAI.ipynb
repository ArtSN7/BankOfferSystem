```python
def item_based_recommendations(user_id, service_matrix, service_similarity_df, num_recommendations=5):
    user_ratings = service_matrix.loc[user_id]
    similar_scores = service_similarity_df.dot(user_ratings).sort_values(ascending=False)
    
    # Убираем уже оцененные пользователем услуги
    similar_recommendations = similar_scores.drop(user_ratings[user_ratings > 0].index)
    
    # Выбираем рекомендации с учетом связей услуг
    final_recommendations = pd.Series(dtype=float)
    
    for service in similar_recommendations.index:
        related_services = cross_services.loc[service]
        related_services = related_services[related_services == 1].index
        if not related_services.empty:
            related_scores = similar_scores[related_scores.index.isin(related_services)]
            final_recommendations = pd.concat([final_recommendations, related_scores])
    
    # Отбираем топ рекомендаций
    final_recommendations = final_recommendations.sort_values(ascending=False).head(num_recommendations)
    
    return final_recommendations
```

```python
from sklearn.metrics import jaccard_score
```

```python
def jaccard_similarity(matrix):
    num_services = matrix.shape[1]
    similarity_matrix = pd.DataFrame(index=matrix.columns, columns=matrix.columns)
    
    for i in range(num_services):
        for j in range(i, num_services):
            service_i = matrix.iloc[:, i]
            service_j = matrix.iloc[:, j]
            similarity = jaccard_score(service_i, service_j)
            similarity_matrix.iloc[i, j] = similarity
            similarity_matrix.iloc[j, i] = similarity
    
    return similarity_matrix
```

```python
def item_based_recommendations(user_id, service_matrix, service_similarity_df, service_relations_df, num_recommendations=5):
    user_interactions = service_matrix.loc[user_id]
    
    # Убедитесь, что индексы совпадают
    if not user_interactions.index.equals(service_similarity_df.index):
        raise ValueError("Индексы user_interactions не совпадают с индексами service_similarity_df")
    
    similar_scores = service_similarity_df.dot(user_interactions).sort_values(ascending=False)
    
    # Убираем уже использованные пользователем услуги
    similar_recommendations = similar_scores.drop(user_interactions[user_interactions > 0].index)
    
    # Выбираем рекомендации с учетом связей услуг
    final_recommendations = pd.Series(dtype=float)
    seen_services = set()
    
    for service in similar_recommendations.index:
        if service not in service_relations_df.index:
            continue
        
        related_services = service_relations_df.loc[service]
        related_services = related_services[related_services == 1].index
        
        if not related_services.empty:
            related_scores = similar_scores[similar_scores.index.isin(related_services)]
            
            # Фильтруем уже виденные услуги для обеспечения разнообразия
            new_scores = related_scores[~related_scores.index.isin(seen_services)]
            final_recommendations = pd.concat([final_recommendations, new_scores])
            
            # Обновляем виденные услуги
            seen_services.update(related_scores.index)
    
    # Убираем дубликаты и отбираем топ рекомендаций
    final_recommendations = final_recommendations.drop_duplicates().sort_values(ascending=False).head(num_recommendations)
    
    return final_recommendations, random.choices(['Экспобанк', 'Д2 страхование', 'Автоэкспресс'], weights=[0.33, 0.5, 0.17])
```
